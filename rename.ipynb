```python
import os
import re

def remove_increment_number(directory):
    # Regex pattern to match filenames with increment numbers at the end
    pattern = re.compile(r"(.*?--\d{8,10})--\d{1,4}\.pdf$", re.IGNORECASE)
    
    for filename in os.listdir(directory):
        if filename.lower().endswith('.pdf'):
            match = pattern.match(filename)
            if match:
                new_filename = match.group(1) + '.pdf'
                old_file_path = os.path.join(directory, filename)
                new_file_path = os.path.join(directory, new_filename)
                
                if old_file_path != new_file_path:
                    if os.path.exists(new_file_path):
                        print(f"Skipping: {new_filename} already exists.")
                    else:
                        os.rename(old_file_path, new_file_path)
                        print(f"Renamed: {filename} -> {new_filename}")

# Set the directory containing the PDF files
pdf_directory = r'E:\1911-1920 (8516)'

remove_increment_number(pdf_directory)

```
